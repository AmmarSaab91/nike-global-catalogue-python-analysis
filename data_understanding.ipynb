```python
print('full_row_duplicates', df.duplicated().sum())
print('duplicate_product_id_sku_pairs', df.duplicated(subset=['product_id', 'sku']).sum())
```

## 3) Measure missing values

```python
global_missing_values = df.isnull().sum()
missing_percent = ((global_missing_values / len(df)) * 100).round(2)

missing_summary = pd.DataFrame({
    'missing_count': global_missing_values,
    'missing_percent': missing_percent
})

print(missing_summary[missing_summary['missing_count'] > 0])
```

```python
sum_of_local_missing_values = pd.Series(0, index=next(iter(dfs.values())).columns)

for data_name, data_frame in dfs.items():
    local_missing_values = data_frame.isnull().sum()
    sum_of_local_missing_values += local_missing_values
    
    print(data_name)
    print(local_missing_values[local_missing_values > 0])
    print()

print("sum_of_local_missing_values:")
print(sum_of_local_missing_values[sum_of_local_missing_values > 0])
```

## 4) Granularity and repetition

This section helps infer the likely row grain by checking repetition in important identifiers.

```python
print('rows', len(df))
print('unique_product_id', df['product_id'].nunique())
print('unique_sku', df['sku'].nunique())
print('unique_product_id_sku_pairs', df[['product_id', 'sku']].drop_duplicates().shape[0])
```

```python
df.groupby(['product_id', 'sku'])['gender_segment'].nunique(dropna=True).value_counts().sort_index()
```

## 5) Main observations

- The dataset contains repeated `product_id` and `sku` values, so the rows are more detailed than simple product-level records.
- Missing values are concentrated in selected descriptive and size-related columns rather than across all fields.
- Duplicate checks should be considered both at the full-row level and at the `product_id` + `sku` level.
- Consistency checks such as `gender_segment` within `product_id` + `sku` help identify which columns may be safe candidates for later cleaning.
- Actual cleaning, imputation, and row removal should be done in a separate notebook.
